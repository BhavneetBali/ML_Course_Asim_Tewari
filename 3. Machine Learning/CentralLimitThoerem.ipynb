# Central Limit Theorem (CLT)

The Central Limit Theorem (CLT) is a fundamental concept in probability and statistics. It states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal (Gaussian) distribution, regardless of the original distribution of the variables.

## Importance of CLT:

- **Universal Applicability:** The CLT applies to a wide range of probability distributions, making it a powerful tool in statistical analysis.
- **Sampling Theory:** The theorem is foundational in the theory of sampling, providing insights into the behavior of sample means.

Now, let's move on to the code implementation and examples.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the original distribution
original_mean = 10
original_std_dev = 2
sample_size = 1000
num_samples = 1000

# Generate samples from the original distribution
original_samples = np.random.normal(original_mean, original_std_dev, size=(num_samples, sample_size))

# Calculate the sample means for each set of samples
sample_means = np.mean(original_samples, axis=1)

# Plot the original distribution and the distribution of sample means
plt.figure(figsize=(12, 6))

# Plot the original distribution
plt.subplot(1, 2, 1)
plt.hist(original_samples.flatten(), bins=50, density=True, color='blue', alpha=0.7)
plt.title('Original Distribution')
plt.xlabel('Values')
plt.ylabel('Probability Density')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, density=True, color='orange', alpha=0.7)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Means')
plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```

## Example Explanation:

In the code above, we generate samples from a normal distribution with a mean of 10 and a standard deviation of 2. We then calculate the sample means for each set of samples and plot both the original distribution and the distribution of sample means.

The left subplot shows the original distribution, while the right subplot illustrates how the distribution of sample means tends to be more Gaussian, supporting the Central Limit Theorem.

Feel free to experiment with different parameters and distributions to observe the impact on the results.


# Use Case: Estimating Population Mean

Consider a scenario where we want to estimate the average height of a population. Measuring the height of every individual is impractical, so we take random samples. According to the Central Limit Theorem, the distribution of sample means will be approximately normal.

We can then use the sample mean to estimate the population mean and calculate a confidence interval.

Let's simulate this scenario with a code example:


```python
# Simulate estimating population mean using CLT
population_mean = 170
population_std_dev = 5
sample_size = 30
num_samples = 1000

# Generate samples from the population
population_samples = np.random.normal(population_mean, population_std_dev, size=(num_samples, sample_size))

# Calculate the sample means for each set of samples
sample_means = np.mean(population_samples, axis=1)

# Calculate the 95% confidence interval for the population mean
confidence_interval = np.percentile(sample_means, [2.5, 97.5])

# Display results
print(f"True Population Mean: {population_mean}")
print(f"Estimated Population Mean: {np.mean(sample_means)}")
print(f"95% Confidence Interval: {confidence_interval}")
```

## Use Case Explanation:

In this use case, we simulate the process of estimating the population mean using the Central Limit Theorem. We assume a true population mean of 170 and a standard deviation of 5. By taking multiple random samples from the population and calculating sample means, we can estimate the population mean and provide a confidence interval.

This demonstrates how the Central Limit Theorem is practically applied in inferential statistics for making reliable estimates about population parameters based on sample data.
